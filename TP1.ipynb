```python
import pandas as pd
import plotly.express as px
import numpy as np

N=2000

D=5

t = np.linspace(-D,D,N)
```

```python
T=2

def porte(t):
    if -T/2<=t<=T/2:
        return 1
    else:
        return 0

s=[]
for i in range (N):
    s.append(porte(t[i]))




df = pd.DataFrame({'Arg1':t, 'Arg2':s})

fig = px.line(df, x='Arg1', y='Arg2')

fig.update_layout(xaxis_title='Argument 1', yaxis_title='Argument 2',
            title = 'Tracé de porte',
            template='plotly_white', width=500, height=300)

fig.show()
```

```python
from msicpe.ssl import TransFourier

S,f = TransFourier(s,t)

df = pd.DataFrame({'Arg1':f, 'Arg2':np.real(S)})

fig = px.line(df, x='Arg1', y='Arg2')

fig.update_layout(xaxis_title='Argument 1', yaxis_title='Argument 2',
            title = 'Tracé de porte',
            template='plotly_white', width=500, height=300)

fig.show()

df = pd.DataFrame({'Arg1':f, 'Arg2':np.imag(S)})

fig = px.line(df, x='Arg1', y='Arg2')

fig.update_layout(xaxis_title='Argument 1', yaxis_title='Argument 2',
            title = 'Tracé de porte',
            template='plotly_white', width=500, height=300)

fig.show()

```
